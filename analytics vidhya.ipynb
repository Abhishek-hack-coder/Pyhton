```python
import pandas as pd 
import numpy as np
data_bm=pd.read_csv("SampleData.csv")
data_bm=data_bm.dropna(how="any")
data_bm=data_bm.reset_index(drop=True)
data_bm.head()
```

```python
status=data_bm.groupby("status")
status.head()
```

```python
status.salary.mean()
```

```python
multiple_groups=data_bm[:10].groupby(["hsc_s","gender"])
multiple_groups.first()
```

```python
t=pd.crosstab(data_bm["hsc_s"],data_bm["gender"],margins=True)
t
```

```python
t.plot(kind='bar', stacked=True, color=['yellow','blue','green'], grid=True)
```
